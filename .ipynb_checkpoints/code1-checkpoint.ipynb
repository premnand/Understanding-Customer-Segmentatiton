### Understanding consumer segmentation for targeted marketing 

```python
# Importing Libraries
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score
from scipy.stats import boxcox
```

```python
# Original Dataset 
df = pd.read_csv('Online_Retail.csv', encoding='windows-1252')
df
```

```python
# Data Preprocessing
# Missing values in all the columns
df.isnull().sum()
```

```python
# Data points with missing CustomerIDs will be excluded from the analysis, as they cannot be attributed to specific customers.
df1 = df.dropna()
df1.isnull().sum()
```

```python
# Dropping duplicate values 
print(f"Number of duplicate rows: {df.duplicated().sum()}")
df1 = df1.drop_duplicates()
```

```python
df1.describe().T
```

```python
# Remove rows with negative Quantity values
df2 = df1[df1['Quantity'] > 0]
```

```python
# Changing datetime to a consistent format
df2['InvoiceDate'] = pd.to_datetime(df2['InvoiceDate'], format='mixed')
```

```python
# Removing cancelled transactions
df2 = df2[~df2['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df2
```

```python
# Histograms for numerical features before outlier removal
df2[['Quantity','UnitPrice']].hist(figsize=(10, 3), bins=20,layout = (1,2), color = 'skyblue',ec="black")
plt.show()
```

```python
# Outlier Removal using IQR
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df3 = df2.copy()
for column in ['Quantity', 'UnitPrice']:
    df3 = remove_outliers(df3, column)

df3
```

```python
# Histograms for numerical features after outlier removal
df3[['Quantity','UnitPrice']].hist(figsize=(10, 3), bins=20,layout = (1,2), color = 'skyblue',ec="black")
plt.tight_layout()
plt.show()
```

```python
# RFM Analysis

# To evaluate the Monetary values at customer level we wil requires the total amount for each purchses
# Creating a 'TotalAmount' column
df3['TotalAmount'] = df3['Quantity'] * df3['UnitPrice']

# For recency will check what was the last date of transaction
latest_date = df3['InvoiceDate'].max() + dt.timedelta(days=1)
```

```python
#RFM factors calculation:
RFM_data = df3.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (latest_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalAmount': 'sum'
}).reset_index()
```

```python
#converting the names of the columns
RFM_data.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalAmount': 'Monetary'
}, inplace=True)
```

```python
RFM_data
```

```python
# RFM_data Summary
RFM_data.iloc[:,1:].describe().T
```

```python
# Visualizing the Recency, Frequency and Monetary distributions
colors = ['red', 'green', 'blue']
columns = ['Recency', 'Frequency', 'Monetary']
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for i, col in enumerate(columns):
    sns.histplot(RFM_data[col], color=colors[i], ax=axes[i])
    axes[i].set_title('Skewness ' + ': ' + round(RFM_data[col].skew(),2).astype(str))

plt.tight_layout()
plt.show()
```

```python
# From the distribution plots above, we can observe that all three metrics are positively skewed (right-skewed).
# This is also confirmed by the skewness coefficient values. To handle the right-skewed distributions, we will apply a log transformation. 

# First will focus on the negative and zero values before the transformation.
def to_one(value):
    return 1 if value <= 0 else value

RFM_data['Monetary'] = RFM_data['Monetary'].apply(to_one)
RFM_data['Recency'] = RFM_data['Recency'].apply(to_one)
RFM_data['Frequency'] = RFM_data['Frequency'].apply(to_one)

# Log Transformation
log_RFM_data = RFM_data[['Recency', 'Frequency', 'Monetary']].apply(np.log1p, axis=1).round(4)
```

```python
# Visualisation After log transformation

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
for i, col in enumerate(columns):
    sns.histplot(log_RFM_data[col], color=colors[i], kde=True, ax=axes[i])
    skewness = round(log_RFM_data[col].skew(), 2)
    axes[i].set_title(f'Skewness: {skewness}')

plt.tight_layout()
plt.show()
```

```python
# Standardization
scaler = StandardScaler()
Scaled_RFM_data = scaler.fit_transform(log_RFM_data)
Scaled_RFM_data = pd.DataFrame(Scaled_RFM_data, index=log_RFM_data.index, columns=log_RFM_data.columns)
```

```python
# Elbow Method to find the optimal number of clusters for customer segmentation
SS_distance = {}
for k in range(1, 20):
    model_with_k = KMeans(n_clusters=k, init='k-means++', max_iter=500)
    model_with_k.fit(Scaled_RFM_data)
    SS_distance[k] = model_with_k.inertia_

plt.figure(figsize=(15, 7))
sns.pointplot(x=list(SS_distance.keys()), y=list(SS_distance.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Sum of square Distances")
plt.title("Elbow Technique to find the optimal cluster size")
plt.show()
```

```python
# K-Means Clustering
KM_model = KMeans(n_clusters=3, init='k-means++', max_iter=500)
KM_model.fit(Scaled_RFM_data)
```

```python
RFM_data['Cluster'] = KM_model.labels_.astype(str)
RFM_data['Cluster'] = 'Cluster #' + RFM_data['Cluster']
```

```python
RFM_data.head()
```

```python
# Visualisation

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Recency vs Frequency
sns.scatterplot(data=RFM_data, x='Recency', y='Frequency', hue='Cluster', ax=axes[0])
axes[0].set_title("Recency vs Frequency")

# Frequency vs Monetary
sns.scatterplot(data=RFM_data, x='Monetary', y='Frequency', hue='Cluster', ax=axes[1])
axes[1].set_title("Frequency vs Monetary")

# Recency vs Monetary
sns.scatterplot(data=RFM_data, x='Monetary', y='Recency', hue='Cluster', ax=axes[2])
axes[2].set_title("Recency vs Monetary")

plt.tight_layout()
plt.show()
```

```python
# Log Visualization
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
sns.scatterplot(data=RFM_data, x='Recency', y='Frequency', hue='Cluster', ax=axes[0])
axes[0].set_title("Recency vs Frequency")
axes[0].set_xscale('log')
axes[0].set_yscale('log')

sns.scatterplot(data=RFM_data, x='Monetary', y='Frequency', hue='Cluster', ax=axes[1])
axes[1].set_title("Frequency vs Monetary")
axes[1].set_xscale('log')
axes[1].set_yscale('log')

sns.scatterplot(data=RFM_data, x='Monetary', y='Recency', hue='Cluster', ax=axes[2])
axes[2].set_title("Recency vs Monetary")
axes[2].set_xscale('log')
axes[2].set_yscale('log')

plt.tight_layout()
plt.show()
```

```python
## KNN Classification
X = RFM_data.drop(['Cluster', 'CustomerID'], axis=1)
y = RFM_data['Cluster']
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
param_grid = {
    'n_neighbors': np.arange(1, 15),
}
knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='accuracy', verbose=1)
grid_search.fit(x_train, y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
```

```python
best_knn = KNeighborsClassifier(**grid_search.best_params_)
best_knn.fit(x_train, y_train)
y_pred = best_knn.predict(x_test)
test_accuracy = best_knn.score(x_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
print("Test Accuracy:", test_accuracy)
```
